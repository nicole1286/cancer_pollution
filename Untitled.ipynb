```python
#import dependencies
import pandas as pd
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#declare csv's and read into pandas DF(Extract data)
csv_file_cancer = "incd_cancer_2016.csv"
csv_file_pollution = "pollution_us_2010.csv"
cancer_df = pd.read_csv(csv_file_cancer, encoding = 'unicode_escape')
pollution_df = pd.read_csv(csv_file_pollution)
```

```python
#Transform data by cleaning into new df's with desired columns only
cancer_clean = cancer_df[['State', 'Incidence Rate- cases per 100,000', 'Average Annual Count', 'Recent Trend', 'Recent 5-Year Trend']]
sorted_cancer = cancer_clean.sort_values(by='Incidence Rate- cases per 100,000', ascending=False)
sorted_cancer.head()
```

```python
pollution_clean = pollution_df[['State', 'NO2 Units', 'NO2 Mean', 'O3 Units', 'O3 Mean', 'SO2 Units', 'SO2 Mean', 'CO Units', 'CO Mean']]
pollution_group = pollution_clean.groupby(['State'])
pollution_state = pollution_group.mean()
pollution_state.head()
```

```python
merge_df = pd.merge(sorted_cancer, pollution_state, how='outer', on=['State'])
merge_df.round(4)
```
